```python
import pandas as pd
import numpy as np
import math 
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Charger les donnes
data = pd.read_csv('dataviz_ready.csv', delimiter=';')

# Convertir la colonne 'departement' en chane de caractres
data['departement'] = data['departement'].astype(str)

# Remplacer les valeurs numriques par des textes
data['evaluation'] = data['evaluation'].replace({0: 'Insalubre', 1: 'A amliorer', 2: 'Satisfaisant', 3: 'Trs satisfaisant'})

# Charger la carte des dpartements franais
map_df = gpd.read_file('https://github.com/gregoiredavid/france-geojson/raw/master/departements-version-simplifiee.geojson')

# Fusionner les donnes avec la carte
merged = map_df.set_index('code').join(data.set_index('departement'))

# Crer la carte thermique
fig, ax = plt.subplots(1, figsize=(10, 6))

# Dfinir la palette de couleurs
cmap = ListedColormap(['red', 'orange', 'lightgreen', 'green'])

# Afficher la carte
merged.plot(column='evaluation', cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, legend_kwds={'bbox_to_anchor': (1, 1)})

# Ajouter un titre
plt.title('valuation sanitaire des tablissements alimentaires des dpartements franais')

plt.show()





# # Charger les donnes
# data = pd.read_csv('dataviz_ready.csv', delimiter=';')


# # Convertir la colonne 'departement' en chane de caractres
# data['departement'] = data['departement'].astype(str)

# # Charger la carte des dpartements franais
# map_df = gpd.read_file('https://github.com/gregoiredavid/france-geojson/raw/master/departements-version-simplifiee.geojson')

# # Fusionner les donnes avec la carte
# merged = map_df.set_index('code').join(data.set_index('departement'))

# # Crer la carte thermique
# fig, ax = plt.subplots(1, figsize=(10, 6))
# merged.plot(column='evaluation', cmap='coolwarm', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# plt.show()

```
