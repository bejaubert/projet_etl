```python
import pandas as pd
import numpy as np
import math 
import geopandas as gpd
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

# Charger les données
data = pd.read_csv('dataviz_ready.csv', delimiter=';')

# Convertir la colonne 'departement' en chaîne de caractères
data['departement'] = data['departement'].astype(str)

# Remplacer les valeurs numériques par des textes
data['evaluation'] = data['evaluation'].replace({0: 'Insalubre', 1: 'A améliorer', 2: 'Satisfaisant', 3: 'Très satisfaisant'})

# Charger la carte des départements français
map_df = gpd.read_file('https://github.com/gregoiredavid/france-geojson/raw/master/departements-version-simplifiee.geojson')

# Fusionner les données avec la carte
merged = map_df.set_index('code').join(data.set_index('departement'))

# Créer la carte thermique
fig, ax = plt.subplots(1, figsize=(10, 6))

# Définir la palette de couleurs
cmap = ListedColormap(['red', 'orange', 'lightgreen', 'green'])

# Afficher la carte
merged.plot(column='evaluation', cmap=cmap, linewidth=0.8, ax=ax, edgecolor='0.8', legend=True, legend_kwds={'bbox_to_anchor': (1, 1)})

# Ajouter un titre
plt.title('Évaluation sanitaire des établissements alimentaires des départements français')

plt.show()





# # Charger les données
# data = pd.read_csv('dataviz_ready.csv', delimiter=';')


# # Convertir la colonne 'departement' en chaîne de caractères
# data['departement'] = data['departement'].astype(str)

# # Charger la carte des départements français
# map_df = gpd.read_file('https://github.com/gregoiredavid/france-geojson/raw/master/departements-version-simplifiee.geojson')

# # Fusionner les données avec la carte
# merged = map_df.set_index('code').join(data.set_index('departement'))

# # Créer la carte thermique
# fig, ax = plt.subplots(1, figsize=(10, 6))
# merged.plot(column='evaluation', cmap='coolwarm', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# plt.show()

```
